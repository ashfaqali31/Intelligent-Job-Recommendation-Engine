```python
postings.drop(['views', 'applies', 'original_listed_time', 'expiry', 'closed_time', 
                   'job_posting_url', 'application_url', 'application_type', 'pay_period', 
                   'max_salary', 'min_salary', 'med_salary', 'company_id', 'remote_allowed',
                   'listed_time', 'posting_domain', 'sponsored', 'currency', 
                   'compensation_type', 'normalized_salary', 'zip_code', 'fips',], axis=1, inplace=True)
```

```python
postings.head()
```

```python
postings.isnull().sum()
```

```python
postings = postings.drop(columns=['skills_desc'])
```

```python
postings = postings.dropna(subset=['description', 'company_name'])
```

```python
postings['formatted_experience_level'] = postings['formatted_experience_level'].fillna('Not Specified')
```

```python
postings.isnull().sum()
```

```python
postings.duplicated().sum()
```

```python
postings['description'] = postings['description'].str.lower().str.strip()
```
