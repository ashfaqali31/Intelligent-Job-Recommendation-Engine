```python
print(f"Total skills in list: {len(skills_list)}")
```

```python
#function to extract skills from resume text
def extract_skills(resume_text, skills_list):
    resume_text = resume_text.lower()
    found_skills = set()
    
    for skill in skills_list:
        pattern = r'\b' + re.escape(skill) + r'\b'
        if re.search(pattern, resume_text):
            found_skills.add(skill)
    return list(found_skills)
```

```python
#test sample
sample_resume = resume['Resume_str'].iloc[0]
extracted_skills = extract_skills(sample_resume, skills_list)
print("Extracted Skills:", extracted_skills)
```

```python
import random

# Pick 10 random resume indices
random_indices = random.sample(range(len(resume)), 10)

for idx in random_indices:
    print(f"\n{'='*60}")
    print(f"RESUME {idx+1} - Category: {resume['Category'].iloc[idx]}")
    print(f"{'='*60}")
    
    sample_resume = resume['Resume_str'].iloc[idx]
    extracted_skills = extract_skills(sample_resume, skills_list)
    
    print(f"Extracted Skills ({len(extracted_skills)}): {extracted_skills}\n")
```
